```python
import chromadb

# Chroma DB uses the all-MiniLM-L6-v2 model from Sentence-Transformers by default , - it is a sentince transformer from hugging face
```

```python
client = chromadb.Client()
client
```

```python
collection = client.create_collection('news_articels')
```

```python
collection.add(
    ids=['id1','id2','id3','id4'],
    documents=[
        'Apple is leading in a smart phone game with iphone sales up by 35%',
        'Tesla booked a minor profit of 1 billion $ in q2',
        'Apples are high in fiber , vitamin c , and various antioxidants',
        'SpaceX got NASA contract worth 10 bbillions $',
    ]

)
```

```python
# Fetch documents + embeddings 

data = collection.get(
    include = ['documents','embeddings','metadatas']
)

# printing embeddings
for i , emb in enumerate(data['embeddings']):
    print(f'\nDocument {i+1}')
    print('Text: ',data['documents'][i])
    print('Embedding length : ', len(emb))
    print('First 10 values : ',emb[:10])
```

```python
results = collection.query(

    query_texts=['this is a query related to Elon Musk'],
    n_results=2
)

print(results)
```

```python
results = collection.query(

    query_texts=['i love apple'],
    n_results=1
)

print(results)
```
