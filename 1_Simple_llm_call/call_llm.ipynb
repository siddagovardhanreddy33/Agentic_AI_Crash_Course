```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model='gemma-4-31b-it')

response = llm.invoke('how many moons does jupiter have ?')
print(response.text)
```

```python
# System instruction 

s_response = llm.invoke([
    ['system','You are a helpful assestant that answer in short onne line'],
    ['human','How many moons does jupiiter have?']
])

print(s_response.text)
```

```python
# Temperature 

prompt = ' write a single tagline for a quirky coffee shop . Only output the tagline , nothing else .'

llm = ChatGoogleGenerativeAI(model = 'gemini-2.5-flash',temperature=0.5)
response = llm.invoke(prompt)
print(response.text)
```

```python

llm = ChatGoogleGenerativeAI(model = 'gemini-2.5-flash',temperature=1)
response = llm.invoke(prompt)
print(response.text)
```

## call llm using Groq Models

```python
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq( model="qwen/qwen3-32b",reasoning_format='parsed')

message = [ 
    (
        "system","You are a helpful assestant that gives a short answer to a question"
    ),

    (
        "human",'How many moons does jupyter have?'
    )
]

ai_msg = llm.invoke(message)

print(ai_msg.content)
```
